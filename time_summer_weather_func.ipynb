```python
matrix_df = coords_df.rename(columns={'SPO_NO_250': 'ID_250'})
```

```python
mon = ['jun', 'jul', 'aug']
weather = ['cloudy', 'sunny','rainy']
```

```python
cloudy_df = pd.read_csv('apr_cloudy.txt')
```

```python
df = pd.DataFrame(columns=cloudy_df.columns)
```

```python
for i in mon:
    for j in weather:
            df3 = pd.read_csv('{mon}_{weather}.txt'.format(mon=i, weather=j))
            df2 = pd.merge(matrix_df, df3)
            df2["month"] = i
            df2["weather"] = j
            
            df = df.append(df2, ignore_index=True)
```

```python
noon_df = df.groupby(['ID_250', 'weather']).agg({'FPC_09T': np.sum, 'FPC_10T': np.sum, 'FPC_11T': np.sum, 'FPC_12T': np.sum, 'FPC_13T': np.sum, \
                                                 'FPC_14T': np.sum, 'FPC_15T': np.sum, 'FPC_16T': np.sum, 'FPC_17T': np.sum, \
                                                 'FPC_18T': np.sum, 'FPC_19T': np.sum, 'FPC_20T': np.sum, 'FPC_21T': np.sum})
noon_df
```

```python
noon_df['morning'] = noon_df['FPC_09T'] + noon_df['FPC_10T'] + noon_df['FPC_11T'] + noon_df['FPC_12T'] 
noon_df['afternoon'] = noon_df['FPC_13T'] + noon_df['FPC_14T'] + noon_df['FPC_15T'] + noon_df['FPC_16T'] + noon_df['FPC_17T']
noon_df['night'] = noon_df['FPC_18T'] + noon_df['FPC_19T'] + noon_df['FPC_20T'] + noon_df['FPC_21T']
```

```python
noon_df = noon_df[['morning', 'afternoon', 'night']]
```

```python
def kmeans_df(c0, c1, c2, c3, c4, c5, c6, c7, c8, c9, c10, c11, c12, c13, c14, c15, c16, c17, c18, c19, c20, c21, c22, c23, c24, c25, c26, \
             c27, c28, c29, c30, c31, c32, c33, c34, c35, c36, c37, c38, c39):
    df = pd.DataFrame(np.hstack([X, 
                                
                                model.labels_[:, np.newaxis]]),
                    columns=["x0", "x1", "c"])
    return df
```

```python
def weather(t, w, r):
    wea_df = pd.DataFrame(noon_df[t])
    
    so_df = wea_df.reset_index('ID_250')
    new_df = so_df.reset_index('weather')
    
    #날씨 select
    weather_df = new_df[new_df['weather'] == w]
    
    merged_df = pd.merge(matrix_df, weather_df)
    # 50 이하의 격좌 제거
    merge_df = merged_df[merged_df['{t}'.format(t=t)] > 50]
    
    #추자도, 제주시 제거
    merg_df = merge_df[((merge_df[merge_df.columns[0]] < 886636)|(merge_df[merge_df.columns[0]] > 893124))|\
                       (merge_df[merge_df.columns[1]] < 1543000)]
    merges_df = merg_df[((merg_df[merg_df.columns[0]] < 902800)|(merg_df[merg_df.columns[0]] > 914500))|\
                       ((merg_df[merg_df.columns[1]] < 1498000)|(merg_df[merg_df.columns[1]] > 1504900))]
    
    list = zip((merges_df[merges_df.columns[0]] - 910228.1215314917), (merges_df[merges_df.columns[1]] - 1488009.814369485))
    X = np.array(list)
    
    global X
    
    from sklearn.cluster import KMeans
    model = KMeans(n_clusters=40, init="random", n_init=1, max_iter=1, random_state=0).fit(X)
    
    global model
    
    c0, c1, c2, c3, c4, c5, c6, c7, c8, c9, c10, c11, c12, c13, c14, c15, c16, c17, c18, c19, c20, c21, c22, c23, c24, c25, c26, \
             c27, c28, c29, c30, c31, c32, c33, c34, c35, c36, c37, c38, c39 = model.cluster_centers_
    cl_df = kmeans_df(c0, c1, c2, c3, c4, c5, c6, c7, c8, c9, c10, c11, c12, c13, c14, c15, c16, c17, c18, c19, c20, c21, c22, c23, c24, c25, c26, \
             c27, c28, c29, c30, c31, c32, c33, c34, c35, c36, c37, c38, c39)
    
    #cl_df와 merges_df merge
    cl_df["x,y"] = zip(cl_df['x0'], cl_df['x1'])
    merges_df['x,y'] = zip((merges_df[merges_df.columns[0]] - 910228.1215314917) , (merges_df[merges_df.columns[1]] - 1488009.814369485))
    mer_df = pd.merge(merges_df, cl_df)
    
    
    pop_df = mer_df.groupby("c").agg({
        '{t}'.format(t=t) : [np.sum,]
    })
    pops_df = pop_df.sort_values(by=pop_df.columns[0], ascending=False)
    
    for i, j in zip(model.cluster_centers_, range(40)):
        if j in [int(pops_df.index[r-1])]:
            plt.scatter(X[model.labels_== j,0], X[model.labels_==j,1], s=100, marker='v', color='b');
            plt.scatter(i[0], i[1], s=100, c='white')
        else:
            plt.scatter(X[model.labels_== j,0], X[model.labels_==j,1], s=100, marker='^', color='white');
            plt.scatter(i[0], i[1], s=100, c='black')
```
